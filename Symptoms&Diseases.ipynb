```python
#Model building
symptoms_disease_models = {
    "Random Forest" : RandomForestClassifier(),
    "Logistic Regression" : LogisticRegression(),
    "K-Nearest Neighbor" : KNeighborsClassifier(),
    "Support Vector Machine" : SVC(),
    "Extreme Gradient Boosting" : XGBClassifier(),
    "Cat Boosting" : CatBoostClassifier(),
    "Light GBM" : LGBMClassifier(),
    "Multinomial Naive Bayes" : MultinomialNB(),
    "Gaussian Naive Bayes" : GaussianNB(),
    "Bernoulli Naive Bayes" : BernoulliNB()
}
```

```python
for name_of_model,model in symptoms_disease_models.items():
  print(model)
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
disease_encoder = LabelEncoder()

#Disease encoding
disease_columns = data_set.Disease
encoded_disease = disease_encoder.fit_transform(disease_columns)


#Symptoms encoding
symptom_encoder = LabelEncoder()
symptom_columns = data_set.drop('Disease',axis=1)
twod_np_symptoms = symptom_columns.values
oned_symptoms = twod_np_symptoms.flatten()
symptom_encoder.fit(oned_symptoms)
encoded_symptoms = symptom_columns.apply(symptom_encoder.transform)


#Join back the disease and symptoms columns
encoded_dataset = encoded_symptoms
encoded_dataset.insert(0,'Disease',encoded_disease)
```

```python
print(encoded_dataset)
```

```python
#Splitting
input_symptoms = encoded_dataset.drop('Disease',axis=1)
output_disease = encoded_dataset.Disease

from sklearn.model_selection import train_test_split
symptom_train,symptom_test,disease_train,disease_test = train_test_split(input_symptoms,output_disease,test_size=0.2)
```

```python
#Training the models
for name_of_model,model in symptoms_disease_models.items():
  model.fit(symptom_train,disease_train)
```

```python
#Prediction
for name_of_model,model in symptoms_disease_models.items():
  prediction = model.predict(symptom_test)
  prediction = disease_encoder.inverse_transform(prediction)
  print(prediction)
```

```python
#Decode the prediction back to categorical data
#prediction = disease
#Predicted diseases have been decoded
#Decode the actual diseases which are disease_test
```

```python
#Accuracy score
#Precision score
list_accuracy = [ ]
list_precision = [ ]
list_recall = [ ]
list_fone = [ ]
list_name = [ ]
import warnings
warnings.filterwarnings("ignore")
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
for name_of_models,model in symptoms_disease_models.items():
  predicted_disease = model.predict(symptom_test)
  predicted_disease = disease_encoder.inverse_transform(predicted_disease)
  actual_disease = disease_encoder.inverse_transform(disease_test)

  #Evaluating the predicted diseases
  accuracy_of_disease = accuracy_score(actual_disease,predicted_disease)
  precision_of_disease = precision_score(actual_disease,predicted_disease,average='macro')
  recall_of_disease = recall_score(actual_disease,predicted_disease,average='macro')
  fone_of_disease = f1_score(actual_disease,predicted_disease,average='macro')
  list_accuracy.append(accuracy_of_disease)
  list_precision.append(precision_of_disease)
  list_recall.append(recall_of_disease)
  list_fone.append(fone_of_disease)
  list_name.append(name_of_models)
metrics_dictionary = {
      "Model" : list_name,
      "Accuracy score" : list_accuracy,
      "Precision score" : list_precision,
      "Recall score" : list_recall,
      "F1 score" : list_fone
  }

df = pd.DataFrame(metrics_dictionary)
print(df)
```
