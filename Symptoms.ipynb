```python
#Import the dataset
import pandas as pd
data_set = pd.read_csv('/content/DiseaseAndSymptoms (1).csv')
print(data_set)
```

```python
#Display all of the column names
names_columns = data_set.columns
list_names_columns = list(names_columns)
print(list_names_columns)
```

```python
#Print only the first row of the dataset
first_row = data_set.iloc[0]
first_row = first_row.dropna()
print(first_row)
```

```python
# To access all of the 10 records of the patient with the Fungal infection disease
fungal_infection_record = data_set.iloc[0:10,0:18]
print(fungal_infection_record)
```

```python
#Rename Nan into Null instead
fungal_infection_record = fungal_infection_record.fillna('None')
print(fungal_infection_record)
```

```python
#How to access the irregularly placed symptoms?
first_symptom = data_set.Symptom_1[0:10]
print(first_symptom.value_counts())
```

```python
#Count the number of each symptom from 1 to 3
f_i = fungal_infection_record.iloc[0:10,1:5]
print(f_i)
```

```python
#To list all of the symptoms of Fungal infection
two_d_fi = f_i.to_numpy().flatten()
print(two_d_fi)
```

```python
#Converting the Numpy array into Pandas series
series_fi = pd.Series(two_d_fi)
print(series_fi)
```

```python
#Now count
count_fi = series_fi.value_counts()
print(count_fi)
```

```python
#Get the symptoms name list
unique_symp = series_fi.unique()
print(unique_symp)
```

```python
length = len(unique_symp)
position = range(length)
print(position)
```

```python
#Get the symptom's name
for name in unique_symp:
  print(name)
```

```python
#Plot the bar chart: Symp. of FI vs Count of Symp.
import matplotlib.pyplot as plt
plt.xlabel('Symptoms of Fungal Infection')
plt.ylabel('Frequency of the symptoms')
width = 0.3
height = count_fi
plt.bar(position,height,width,color='Blue')
plt.xticks(position,labels=unique_symp,rotation=45)
plt.show()
```

```python
#The Pandas series that got converted into a Numpy array by unique() should be converted into a list
l = list(unique_symp)
l.remove('None')
print(l)
```

```python
#Count the number of symptoms in a list
a = len(l)
print(f"There are {a} symptoms of Fungal infection")
```

```python
#Grouped all of the 41 diseases' dataframe
hehe = data_set.groupby(data_set.Disease)
print(hehe)
```

```python
#Tested to see if each disease got their dataframe or not
hehe_test = hehe.get_group('AIDS')
hehe_test = hehe_test.fillna('None')
print(hehe_test)
```

```python
for name,data_frame in data_set.groupby('Disease'):
  data_frame = hehe.get_group(name)
  data_frame = data_frame.fillna('None')
  print(data_frame)
```

```python
#Design a function to return the count of the symptoms
def work_please(b):
  b = b.iloc[:,1:18]
  one_d_disease = b.to_numpy().flatten()
  series_disease = pd.Series(one_d_disease)
  unique_symptoms = series_disease.unique()
  ls = list(unique_symptoms)
  ls.remove('None')
  finally_count = len(ls)
  return finally_count

rog_ko_naam = data_frame['Disease'].unique()
no_of_symptoms = work_please(data_frame)
print(f"The {rog_ko_naam} has {no_of_symptoms} symptoms.")
```

```python

for name_two, data_frame_two in data_set.groupby('Disease'):
  def work_please(b):
    b = b.iloc[:,1:18]
    b = b.fillna('None')
    one_d_disease = b.to_numpy().flatten()
    series_disease = pd.Series(one_d_disease)
    unique_symptoms = series_disease.unique()
    ls = list(unique_symptoms)
    ls.remove('None')
    finally_count = len(ls)
    return finally_count

  rog_ko_naam = data_frame_two['Disease'].unique()
  no_of_symptoms = work_please(data_frame_two)
  print(f"The {rog_ko_naam[0]} has {no_of_symptoms} symptoms.")
```

```python
naam_list = [ ]
symptoms_list = [ ]
for name_two, data_frame_two in data_set.groupby('Disease'):
  def work_please(b):
    b = b.iloc[:,1:18]
    b = b.fillna('None')
    one_d_disease = b.to_numpy().flatten()
    series_disease = pd.Series(one_d_disease)
    unique_symptoms = series_disease.unique()
    ls = list(unique_symptoms)
    ls.remove('None')
    finally_count = len(ls)
    return finally_count

  rog_ko_naam = data_frame_two['Disease'].unique()
  no_of_symptoms = work_please(data_frame_two)
  naam_list.append(rog_ko_naam[0])
  symptoms_list.append(no_of_symptoms)
print("Disease name:",naam_list)
print("No. of symptoms:",symptoms_list)
```

```python
minimum_symptoms = [ ]
maximum_symptoms = [ ]
for i in range(0,41):
  if(symptoms_list[i] > 7):
    maximum_symptoms.append(naam_list[i])
  elif(symptoms_list[i] < 7 or symptoms_list[i] == 7):
    minimum_symptoms.append(naam_list[i])
print(minimum_symptoms)
print(maximum_symptoms)
```
